```python
import pandas as pd
```

```python
df = pd.read_csv('LondonCOB.csv')
df
```

```python
df_london = df.loc[df.Area.apply(lambda x: x in {'Barking and Dagenham', 'Barnet', 'Bexley', 'Brent', 
               'Bromley', 'Camden', 'City of London,Westminster', 'Croydon',
               'Ealing', 'Enfield', 'Greenwich', 'Hackney', 'Hammersmith and Fulham',
               'Haringey', 'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington',
               'Kensington and Chelsea', 'Kingston upon Thames', 'Lambeth', 
               'Lewisham', 'Merton', 'Newham', 'Redbridge', 'Richmond upon Thames', 
               'Southwark', 'Sutton', 'Tower Hamlets', 'Waltham Forest', 'Wandsworth' })].copy()
df_london
```

```python
totals = df_london.groupby('Area')['Usual Residents'].sum()
totals
```

```python
df_london
#df_london['proportion'] = df_london.apply(lambda row: df_london['Usual Residents'][row]/totals[df_london['Area'][row]])
df_london_location = df_london.copy()
df_london_location['totals'] = df_london_location.groupby('Area')['Usual Residents'].transform('sum')
df_london_location['proportion'] = df_london_location['Usual Residents']/df_london_location.totals
df_london_location
```

```python
df_lambda = df_london.copy()
df_lambda_grouped = df_lambda.groupby(['Area', df_lambda.index]).agg({'Usual Residents': 'sum'})
percents = df_lambda_grouped.groupby(level=0).apply(lambda x: x / float(x.sum()))
df_lambda_grouped
percents = percents.rename(columns={'Usual Residents': 'proportion'})
#percents = percents.reset_index(drop=True)
percents
```

This has grouped by Area, so it has found the total number of people by area. We have then found the proportion of ages that live in each specific area. So within Barking and Dagenham, 10% of the population are between ages 0 to 4. 

```python
df_a_a_london = df_london_location.groupby(['Area','Age'])['proportion'].sum()
df_a_a_london
```

```python
df_a_a_london2 = df_london_location.pivot_table(values='proportion', index='Area', columns='Age', aggfunc='sum')
df_a_a_london2
```

```python
df_age_totals = df_london.groupby('Age')['Usual Residents'].sum()
df_age_totals
```

This is slightly different to the other pivot table above, here we have found the total number of each age range. We have then found the proportion of each age range in regards to the total age range. In the pivot table 2 cells below, we have found for example that 3% of all ages 0 to 4 live in Barking and Dagenham. The 'max' column tells us which age range has the highest value per borough.

```python
df_london_age = df_london.copy()
df_london_age['total'] = df_london_age.groupby('Age')['Usual Residents'].transform('sum')
df_london_age['proportion'] = df_london_age['Usual Residents']/df_london_age['total']
df_london_age = df_london_age.drop(columns=['Sex', 'BirthCountry', 'BirthRegion'])
df_london_age
```

```python
df_a_l = df_london_age.pivot_table(values='proportion', columns='Age', index='Area', aggfunc='sum')

df_a_l['max'] = df_a_l.idxmax(axis=1)
df_a_l
```

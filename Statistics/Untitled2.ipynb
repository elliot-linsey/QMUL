```python
q = pd.DataFrame({
    'Name': ['Abi', 'Ben', 'Charlie', 'Dan'],
    'Mark1': [10,9,7,8],
    'Mark2': [6,7,10,10]
})
q
```

```python
def grade(m):
    if m >= 9:
        return 'A'
    elif m >= 7: 
        return 'B'
    return 'C'
```

```python
q_1 = q.assign(G1 = q.apply(lambda r: grade(r.Mark1), axis=1))
q_1
```

```python
q_2 = q.assign(G1 = q.Mark1.apply(grade), G2 = q.Mark2.apply(grade)) 
q_2
```

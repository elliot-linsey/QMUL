# Hypothesis testing using t-test

The purpose of this notebook is to show how to code a t-test using scipy.stats. Sample data is randomly sampled, so the answer varies on each run. The notebook looks at:
 * The relationship between the sample and population mean: how to calculate a confidence interval on the sample mean.
 * The difference between two means: how to calculate p-value for the hypothesis that the means are different.

The functions introduced are:
 * `stats.t.interval`
 * `stats.ttest_ind`

   2. Random Data Generation
   3. Histograms of the Generated Data

   1. Calculation of the Confidence Interval using Student's t
   2. Calculation of the Confidence Interval using a Normal Approximation

3. Section 3: Hypothesis Test for the Difference between Two Means
   1. Boxplot of Distributions being Compared

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

## Section 1: Set-up

### Section 1.1: A Plot of Student's t-Distribution Plotted
The `scipy.stats` library has the distribution `t`. The distribution has a single parameter, the 'degrees of freedom' (df). The plot shows the distribution and the normal distribution: as the 'degrees of freedom' increases the t-distribution approaches a normal.

```python
dfs =[1, 3, 10]
fig, ax = plt.subplots(1, 1, figsize=(12,6))

# Display the probability density function (pdf):

x = np.linspace(stats.t.ppf(0.01, 2),
                stats.t.ppf(0.99, 2), 100)
ax.plot(x, stats.t.pdf(x, dfs[0]),'red', lw=2, alpha=0.6, label='t df=%d' % dfs[0])
ax.plot(x, stats.t.pdf(x, dfs[1]),'green', lw=2, alpha=0.6, label='t df=%d' % dfs[1])
ax.plot(x, stats.t.pdf(x, dfs[2]),'blue', lw=2, alpha=0.6, label='t df=%d' % dfs[2])
ax.plot(x, stats.norm.pdf(x),'black', lw=2, alpha=0.6, label='norm(0,1)')

ax.legend()
```

### Section 1.2: Random Data Sampled from Normal Distributions

The following code generates data in three classes: A, B and C. There are 200 data points split between the three types in unequal proportion and sampled from different normal distributions. 


| Type | Proportion | Sampled from |
|:----:|:-------------:|:---------|
| A | 40% | `norm(0,1)` |
| B | 35% | `norm(0.35, 1)` |
| C | 25% | `norm(0.7, 1.7)` |

The final data frame shows the 'type' and the value. 

```python
# generate a normal sample
Z = 200

df = pd.DataFrame(data={'R': np.random.random_sample(Z),
                        'A': stats.norm.rvs(size=Z),
                        'B': stats.norm.rvs(loc = 0.35, size=Z),
                        'C': stats.norm.rvs(loc = 0.7, scale = 1.7, size=Z)})
df = df.assign(Type = df.apply(lambda r : 'A' if r.R < 0.4 else ('B' if r.R < 0.75  else 'C'), axis=1))
df = df.assign(Value = df.apply(lambda r: r.A if r.Type =='A' else (r.B if r.Type =='B' else r.C), axis=1))
df = df.loc[:,['Type','Value']]
df
```

### Section 1.3: Histograms of the Generated Data

The following plot shows the distribution of the randomly generated data along with the sample statistics. 

```python
xA = df.loc[df.Type=='A']
xB = df.loc[df.Type=='B']
xC = df.loc[df.Type=='C']

a_mean = xA.mean()
b_mean = xB.mean()
c_mean = xC.mean()

a_std = xA.std(ddof=1)
b_std = xB.std(ddof=1)
c_std = xC.std(ddof=1)

fig, ax = plt.subplots(1,1, figsize=(10,10))
xA.hist(alpha=0.3, ax=ax, label=r'A: $n = %d, \bar{x} = %3.2f, s = %3.2f$' % (len(xA), a_mean, a_std))
xB.hist(alpha=0.3, ax=ax, label=r'B: $n = %d, \bar{x} = %3.2f, s = %3.2f$' % (len(xB), b_mean, b_std))
xC.hist(alpha=0.3, ax=ax, label=r'C: $n = %d, \bar{x} = %3.2f, s = %3.2f$' % (len(xC), c_mean, c_std))
ax.legend(fontsize=14)
```

## Section 2: Confidence Interval for the Mean
We have a sample mean and we want to know what we reliably know about the population means. One approach is to calculate a confidence interval (CI). 
 * The CI is calculated for a significance - 90%, 95% or 99%.
 * It relates to the range in the sample distribution that contains 90% (or 95% or 99%) of density.
 * The sample distribution is the t-distribution.

### Section 2.1: Calculation of the Confidence Interval Using Student's t

The interval is:
$$
   \bar{x} \pm t_{n-1}(\alpha) . \frac{s}{\sqrt{n}}
$$
where
 * $n$ is the size of the sample
 * The degrees of freedom is $n - 1$
 * $\bar{x}$ is the sample mean
 * $s$ is the standard deviation calculated from the sample 
 * $\alpha$ is the level of significance

In Python, this is calculated as: `stats.t.interval(`$\alpha$ `,` $n-1$ `, loc=` $\bar{x}$ `, scale=` $\frac{s}{\sqrt{n}}$ `)`

Remember that the interpretation of the 90% (e.g.) CI is that if we repeat the sampling process many times the population mean will lie with the 90% CI calculated from the sample mean 9 out of 10 times. The other 1 out 10 times, the sample mean is so far from the population mean (just bad luck) that the population mean will lie outside the range calculated. 

```python
lbA9, ubA9 = stats.t.interval(0.99, len(xA)-1, loc=a_mean, scale=a_std / np.sqrt(len(xA)))
print('99%% confidence interval for mean of A is %3.2f to %3.2f' % (lbA9[0], ubA9[0]))

lbA5, ubA5 = stats.t.interval(0.95, len(xA)-1, loc=a_mean, scale=a_std / np.sqrt(len(xA)))
print('95%% confidence interval for mean of A is %3.2f to %3.2f' % (lbA5[0], ubA5[0]))

lbA0, ubA0 = stats.t.interval(0.90, len(xA)-1, loc=a_mean, scale=a_std / np.sqrt(len(xA)))
print('90%% confidence interval for mean of A is %3.2f to %3.2f' % (lbA0[0], ubA0[0]))

lbB, ubB = stats.t.interval(0.95, len(xB)-1, loc=b_mean, scale=b_std / np.sqrt(len(xB)))
print('95%% confidence interval for mean of B is %3.2f to %3.2f' % (lbB[0], ubB[0]))

lbC, ubC = stats.t.interval(0.95, len(xC)-1, loc=c_mean, scale=c_std / np.sqrt(len(xC)))
print('95%% confidence interval for mean of C is %3.2f to %3.2f' % (lbC[0], ubC[0]))
```

### Section 2.2: Calculation of the Confidence Interval Using Normal Approximation

Rather than using the t distribution, the normal can be used. To do this we use `stats.norm.interval` with the same parameters except that the degrees of freedom is not needed.

The answers are very close. This is because our sample size is quite large, so the Student t-distribution is very close to normal. The 99% interval is (often) slightly different, as the tails of the t-distribution have more influence on this than on the 90% interval. Note also that type A is the larger of the three samples, with approximately 80 data values. 

```python
lbA_n9, ubA_n9 = stats.norm.interval(0.99, loc=a_mean, scale=a_std / np.sqrt(len(xA)))
print('99%% confidence interval for mean of A is %3.2f to %3.2f' % (lbA_n9[0], ubA_n9[0]))

lbA_n5, ubA_n5 = stats.norm.interval(0.95, loc=a_mean, scale=a_std / np.sqrt(len(xA)))
print('95%% confidence interval for mean of A is %3.2f to %3.2f' % (lbA_n5[0], ubA_n5[0]))

lbA_n0, ubA_n0 = stats.norm.interval(0.90, loc=a_mean, scale=a_std / np.sqrt(len(xA)))
print('90%% confidence interval for mean of A is %3.2f to %3.2f' % (lbA_n0[0], ubA_n0[0]))
```

### Section 2.3: Plotting a CI on a Histogram

We can show the confidence interval on a plot of the distribution. Note that these lines do NOT contain 90% or 99% of the distribution of the variable (here it is A) itself. Rather 90% (or 99%) of the distribution of the **sample mean** is in this range. We can see that the sample mean is indeed in the centre of the lines. 

If you replace the type A data with data from type C then expect that the interval will be wider as the size of the sample is smaller. 

```python
fig, ax = plt.subplots(1,1, figsize=(8,8))
xA.hist(alpha=0.3, ax=ax, label=r'$n = %d, \bar{x} = %3.2f, s = %3.2f$' % (len(xA), a_mean, a_std))
ax.axvline(lbA_n0[0], c='black', ls='--', label='90% CI for mean')
ax.axvline(ubA_n0[0], c='black', ls='--')
ax.axvline(lbA_n9[0], c='red', ls='--', label='99% CI for mean')
ax.axvline(ubA_n9[0], c='red', ls='--')
ax.legend(fontsize=14)
```

## Section 3: Hypothesis Test for Difference between Two Means

We have two sample of data for different individuals (e.g. boys and girls) and wish to know whether the mean values calculated from the sample are reliably different. The hypotheses are:
 * The null hypothesis ($H_0$) is that the true means could be the same.
 * The alternative hypothesis ($H_1$) is that they are really different.

There are two variants of the test:
  * The paired test, when the individuals are the same and the the sample values are paired. An example is when the data is the average score in Maths and in English for the same class.
  * The unpaired test, when the score are for different individuals. An example would be the English score for the boys in the class against the same score for the girls. There may be a different number of individuals in each group so the data are not paired. 

### Section 3.1: Boxplot of the Distributions being Compared
The boxplot below shows that the the 'B' data appears to have a greater mean than the 'A' data (that is, the sample mean for type B exceeds that for type A).

**One Way or Two Way Test**
Are we concerned about whether the mean of type B is greater than that of A or merely about whether it is different? The distinction is between a one way and two way test (alternative names are *one tail* and *two tail*). The hypotheses for a two way test are given above, those for a one way test are:
 * The null hypothesis ($H_0$) is that the true mean of type B is could be the same or less than that of group A.
 * The alternative hypothesis ($H_1$) is that the the true mean of type B really is the greater. 

```python
a = df.boxplot(column='Value', by='Type', figsize=(14,6))
a.set_title('')
```

### Section 3.2: t-Test for Difference between two Means (Not Paired - Welch's t-test)

The test to compare two means from samples of different size is sometimes called *Welch's test*. The Python function `stats.ttest_ind` calculates this directly from the data. The parameter `equal_var` is `True` when we can assume that the variance of the two sample datasets is the same. Note that for a paired test we use a different function `ttest_rel`.  

```python
resAB = stats.ttest_ind(df.loc[df.Type=='A'].Value, df.loc[df.Type=='B'].Value, equal_var=False )
# from book p111

print('Compare A (mean %3.2F) and B (mean %3.2F): p-value for a one tail test = %5.4f' % 
      (a_mean, b_mean, resAB.pvalue / 2))

resAC = stats.ttest_ind(df.loc[df.Type=='A'].Value, df.loc[df.Type=='C'].Value, equal_var=False )
print('Compare A (mean %3.2F) and C (mean %3.2F): p-value for a one tail test = %5.4f' % 
      (a_mean, c_mean, resAC.pvalue / 2))

resBC = stats.ttest_ind(df.loc[df.Type=='B'].Value, df.loc[df.Type=='C'].Value, equal_var=False )
print('Compare B (mean %3.2F) and C (mean %3.2F): p-value for a one tail test = %5.4f' % 
      (b_mean, c_mean, resBC.pvalue / 2))
```

The function `stats.ttest_ind` returns the tests statistic and the p-value, assuming that a two tailed test is being done. To get the p-value for the single tailed test, we just divide by two. 

### Section 3.3: Accepting or Rejecting a Null Hypothesis 

We can accept or reject the null hypothesis by comparing the calculated p-value with the threshold we have chosen. When doing a single-tailed test remember to check that the sample means are in the correct relationship. 

Here is some sample code that prints a message.

```python
def hypo_test_single(res, alpha, var1, var2):
    if (res.pvalue / 2) < (1 - alpha):
        print('Reject the null hypothsis at %d%% significance: mean of %s greater than mean of %s' 
              % (100 * alpha, var1, var2))
    else: 
        print('Accept the null hypothsis at %d%% significance: mean of %s may not be greater than mean of %s' 
              % (100 * alpha,var1, var2))
        
hypo_test_single(resAB, 0.99, 'B', 'A')
hypo_test_single(resAC, 0.95, 'C', 'A')
```

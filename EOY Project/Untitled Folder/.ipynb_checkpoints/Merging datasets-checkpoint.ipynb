```python
import pandas as pd
import glob
import os
import shutil
import re
```

```python
src_folder = r"D:\EOY Datasets\South West"
```

```python
df1 = pd.read_csv(src_folder+'\\'+'South_West_street.csv',usecols=['Crime ID', 'Month', 'Reported by','Crime type'])
df2 = pd.read_csv(src_folder+'\\'+'South_West_outcomes.csv',usecols=['Crime ID','Outcome type'])
```

```python
df1.head()
```

```python
df1 = df1.dropna(subset=['Crime ID']).reset_index(drop=True)
df2 = df2.dropna(subset=['Crime ID']).reset_index(drop=True)
```

```python
df3 = df1.merge(df2, how='inner',on='Crime ID')
```

```python
df3 = df3.drop_duplicates(subset=['Crime ID'])
```

```python
import pandas as pd
```

```python
se = pd.read_csv(r"D:\EOY Datasets\South_East_full.csv",index_col=0)
```

```python
se01 = se[se.Month=='2019-01'].copy()
```

```python
se01['Region'] = 'South East'
```

```python
se01 = se01.rename(columns={'Crime ID':'Crime_ID', 'Reported by': 'Reported_by', 'Crime type': 'Crime_type', 'Outcome type':'Outcome_type'})
```

```python
se01['Year'] = 2019
```

```python
se01['Level_of_Lockdown'] = 'No Lockdown'
```

```python
se01['Crime_type'] = se01['Crime_type'].apply(lambda x: 'Theft offences' if x in 
                                              ['Other theft', 'Burglary', 'Shoplifting','Theft from the person','Robbery',
                                              'Bicycle theft'] else x)
```

```python
se01 = se01[se01['Crime_type']!='Vehicle crime'].copy()
```

```python
se01['Crime_type'] = se01['Crime_type'].apply(lambda x: 'Public order offences' if x=='Public order' else x)
```

```python
se01 = se01[se01['Crime_type']!='Other crime']
se01 = se01[se01['Crime_type']!='Possession of weapons']
```

```python
se01['Crime_type'] = se01['Crime_type'].apply(lambda x: 'Drug offences' if x=='Drugs' else x)
```

```python
se01['Crime_type'].value_counts()
```

```python
se01['Outcome_type'] = se01['Outcome_type'].apply(lambda x: 'No suspect identified' if x=='Investigation complete; no suspect identified'
                                                  else ('Unable to prosecute suspect' if x == 'Unable to prosecute suspect' else ('Suspect charged' if x == 'Suspect charged'
                                                 else 'Other Outcome')))
```

```python
se01['Outcome_type'].value_counts()
```

```python
se01['Reported_by'] = se01['Reported_by'].str.replace('Service','',regex=True)
se01['Reported_by'] = se01['Reported_by'].str.replace('Police','',regex=True)
se01['Reported_by'] = se01['Reported_by'].str.replace('Constabulary','',regex=True)
se01['Reported_by'] = se01['Reported_by'].str.strip()
```

```python
se01['Reported_by'].value_counts()
```

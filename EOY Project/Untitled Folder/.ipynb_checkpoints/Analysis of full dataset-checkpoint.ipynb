```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"D:\EOY Datasets\Full Datasets"+'\\'+'Full_dataset.csv',index_col=[0])
```

```python
df.head()
```

```python
grouped = df.groupby(['Year','Outcome type'])['Crime ID'].count().reset_index()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=grouped,x='Year',y='Crime ID',hue='Outcome type')
plt.show()
```

```python
grouped2 = df.groupby(['Crime type','Outcome type'])['Crime ID'].count().reset_index()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=grouped2,x='Crime type',y='Crime ID',hue='Outcome type')
plt.show()
```

```python
grouped3 = df.groupby(['Year','Crime type'])['Crime ID'].count().reset_index()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=grouped3,x='Year',y='Crime ID',hue='Crime type')
plt.show()
```

```python
grouped4 = df.groupby(['Year','Crime type','Outcome type'])['Crime ID'].count().reset_index()
grouped4
```

```python
pivot = pd.pivot_table(grouped4,values='Crime ID',columns=['Year','Crime type'],index='Outcome type')
```

```python
pivot
```

Normalised by column, e.g in 2019, for every criminal damage case 74% had no suspect identified, 4% had an other outcome etc. 

```python
pivot2 = pivot/pivot.sum()
pivot2.applymap(lambda x: "{0:.0f}%".format(100*x))
```

```python
pivot3 = pivot/pivot.sum()*100
pivot3
```

```python
pivot4 = pd.pivot_table(grouped4,values='Crime_ID',columns=['Year','Outcome_type'],index='Crime_type')
pivot4
```

```python
pivot2019 = pivot4.loc[slice(None),(2019)]
#pivot2019 = pivot2019/pivot2019.sum()
pivot2019 = pivot2019.div(pivot2019.sum(axis=1), axis=0)
pivot2019
```

```python
tab2 = pd.crosstab(columns=[df['Year'],df['Crime_type']],index=df['Outcome_type'])
tab2
```

```python
def pivoting_stacks(df,year):
    res = df.loc[slice(None),(year)]
    res = res.div(res.sum(axis=1), axis=0)
    return res
```

```python
pivot2019 = pivoting_stacks(pivot4,2019)
pivot2020 = pivoting_stacks(pivot4,2020)
pivot2021 = pivoting_stacks(pivot4,2021)
```

```python
fig, (a1,a2,a3) = plt.subplots(3,1,figsize=(10,14))
pivot2019.plot.bar(stacked=True,ax=a1,title='2019').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
pivot2020.plot.bar(stacked=True,ax=a2,title='2020').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
pivot2021.plot.bar(stacked=True,ax=a3,sharex=a1,rot=0,title='2021').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
tab3 = pd.crosstab(columns=[df['Year'],df['Outcome_type']],index=df['Region'])
tab3
```

```python
pivot2019_2 = pivoting_stacks(tab3,2019)
pivot2020_2 = pivoting_stacks(tab3,2020)
pivot2021_2 = pivoting_stacks(tab3,2021)
```

```python
fig, (a1,a2,a3) = plt.subplots(3,1,figsize=(10,14))
pivot2019_2.plot.bar(stacked=True,ax=a1,title='2019').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
pivot2020_2.plot.bar(stacked=True,ax=a2,title='2020').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
pivot2021_2.plot.bar(stacked=True,ax=a3,sharex=a1,rot=0,title='2021').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
plt.show()
```

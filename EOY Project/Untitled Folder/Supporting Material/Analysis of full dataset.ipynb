Testing the differences between using pd.pivot_table and pd.crosstab. In this case, both are sufficient. 

```python
pivot4 = pd.pivot_table(grouped4,values='Crime_ID',columns=['Year','Outcome_type'],index='Crime_type')
pivot4
```

```python
pivot2019 = pivot4.loc[slice(None),(2019)]
#pivot2019 = pivot2019/pivot2019.sum()
pivot2019 = pivot2019.div(pivot2019.sum(axis=1), axis=0)
pivot2019
```

```python
tab2 = pd.crosstab(columns=[df['Year'],df['Crime_type']],index=df['Outcome_type'])
tab2
```

Creating a function that does the correct slicing and converting to probabilities. 

```python
def pivoting_stacks(df,year):
    res = df.loc[slice(None),(year)]
    res = res.div(res.sum(axis=1), axis=0)
    return res
```

```python
pivot2019 = pivoting_stacks(pivot4,2019)
pivot2020 = pivoting_stacks(pivot4,2020)
pivot2021 = pivoting_stacks(pivot4,2021)
```

```python
plt.rcParams['font.size'] = '10'
```

Plotting the probability of an outcome by crime type.

```python
fig, (a1,a2,a3) = plt.subplots(3,1,figsize=(11,14),constrained_layout = True)
fig.suptitle('Probability of Outcome by Crime',x=0.39,y=1.02, fontsize=16)
p1 = pivot2019.plot.bar(stacked=True,ax=a1,rot=0,title='2019',colormap='icefire',ylabel='Probability').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
p2 = pivot2020.plot.bar(stacked=True,ax=a2,rot=0,title='2020',colormap='icefire',ylabel='Probability').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
p3 = pivot2021.plot.bar(stacked=True,ax=a3,rot=0,title='2021',colormap='icefire',ylabel='Probability').legend(loc='center left',bbox_to_anchor=(1.0, 0.5))

for container in a1.containers:
    a1.bar_label(container,label_type='center')
for container in a2.containers:
    a2.bar_label(container,label_type='center')
for container in a3.containers:
    a3.bar_label(container,label_type='center')
fig.savefig('Probability of outcome by crime.png', dpi=fig.dpi, bbox_inches='tight')
plt.show()
```

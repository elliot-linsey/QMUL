```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tablewithregions.csv')
df.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
```

```python
df
```

```python
df['one'] = pd.cut(df.Number_of_Arrests,bins=np.arange(0,220,20),include_lowest=True)
df.one.value_counts()
```

```python
df['two'] = pd.cut(df.Number_of_Arrests,bins=np.arange(200,2100,100),include_lowest=False)
df.two.value_counts()
```

```python
df['three'] = df.Number_of_Arrests.apply(lambda x: 'Over 2000' if x > 2000 else 'nan')
#df.three.value_counts()
df[df.three == 'Over 2000']
```

```python
df.one, df.two, df.three = df.one.astype('str'),df.two.astype('str'),df.three.astype('str')
```

```python
df['test'] = df.one + df.two + df.three
df
```

```python
df.test = df.test.astype('str').str.replace(',','.',regex=True)
df.test = df.test.str.replace('nan','',regex=True)
```

```python
df.test.value_counts()
```

```python
df.drop(columns=['one','two','three','Arrests_Ranked'],axis=1,inplace=True)
df.rename(columns={'test':'Arrests_Ranked'},inplace=True)
df
```

```python
#df.Arrests_Ranked.value_counts()
```

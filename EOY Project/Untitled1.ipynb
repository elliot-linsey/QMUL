```python
df3 = df.copy()
df3.Police_Force_Name.unique()
```

```python
df3.drop(labels=df3.columns[[0]],axis=1,inplace=True)
df3['Region'] = df3.Police_Force_Name.apply(lambda x: 'South East' if x in ['Essex','Kent','London. City of','Metropolitan Police','Hertfordshire',
                                                                           'Bedfordshire','Surrey','Hampshire','Thames Valley','Sussex'] else 'South West' if (x in ['Devon and Cornwall',
                                                                           'Dorset','Avon and Somerset','Wiltshire','Gloucestershire','Gwent','South Wales',
                                                                           'Dyfed-Powys','West Mercia','North Wales']) else 'West Midlands' if (x in ['Warwickshire','West Midlands',
                                                                            'Staffordshire','Derbyshire','Cheshire','Greater Manchester','Merseyside','Lancashire']) else 'East Midlands' 
                                                                            if (x in ['Suffolk','Cambridgeshire','Norfolk','Northamptonshire','Leicestershire','Lincolnshire','Nottinghamshire',
                                                                                    'South Yorkshire','Humberside','West Yorkshire']) else 'North' if (x in ['North Yorkshire','Durham',
                                                                                    'Northumbria','Cumbria','Cleveland']) else 'British Transport Police' if (x == 'British Transport Police') else 'butts')
df3
```

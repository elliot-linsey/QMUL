```python
df4[df4.Number_of_Arrests == 1987]
```

```python
df4.Arrests_Ranked = df4.apply(lambda row: 'Over 2000' if row['Number_of_Arrests'] > 2000 else row['Arrests_Ranked'],axis=1)
```

```python
df4.Arrests_Ranked = df4.Arrests_Ranked.astype('str').str.replace('nan','',regex=True)
```

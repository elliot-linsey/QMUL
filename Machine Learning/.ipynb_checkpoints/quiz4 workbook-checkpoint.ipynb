```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.metrics import mean_squared_error
np.set_printoptions(precision=3)
```

```python
xTrain = np.load('quiz4xTrain.npy')
yTrain = np.load('quiz4yTrain.npy')
xVal = np.load('quiz4xVal.npy')
yVal = np.load('quiz4yVal.npy')
```

```python
print(xTrain)
print(yTrain)
print(xVal)
print(yVal)
```

```python
x_lin = np.linspace(-3,3,100)
```

```python
p = np.poly1d(np.polyfit(xTrain, yTrain, 6))
y_hat = p(xTrain)
print(mean_squared_error(yTrain,y_hat))
yPredLin = p(x_lin)
yPred6 = p(-2.14)
print(yPred6)
plt.scatter(xTrain,yTrain)
plt.plot(x_lin, yPredLin)
plt.show()
```

Using the above information we can create a tree diagram. At this stage, ignore the TPR and FPR values, they do not come into the tree diagram. The only information needed is the priors and the likelihoods. The prior of population with symptoms is given in part a, the likelihoods are the numbers of people who have symptoms that also have the virus, and those that don't have symptoms that have the virus. 

5% of the population have symptoms, so our first branch is this value of 0.05 and 0.95 (for those that don't have symptoms). From here, the likelihoods are as follows. That one in five who have symptoms also have the virus so P(S|V) = 0.2. Therefore, P(S|notV) must be 0.8.

For the other branch, 1 in 2000 who don't have symptoms have the virus which is 0.0005 so not having symptoms and not having the virus must equal 0.9995.

a) We just sum the paths that have V = True, these are 0.05 * 0.2 + 0.95 * 0.0005 = 0.010475. **Remember these are probabilities, to get percentages you must multiply by 100** so the answer is **1.0475%**

```python
(0.05 * 0.2 + 0.95 * 0.0005)*100
```

b) This utilises the TPR and FPR. Our TPR is 0.9 and FPR is 0.05 for people with symptoms. We just multiply the TPR value by our positive likelihood and our FPR by our negative likelihood = 0.9 * 0.2 + 0.05 * 0.8 = 0.22 or 22%

Our TPR is the chance that we correctly label a positive as positive, in this scenario there is a 0.2 probability that you have the symptoms given you have the virus, as you have the virus there is a 0.9 probability that the test will correctly register you having the virus.

On the other hand, the FPR is the probability that you incorrectly classify what should be negative as a positive. In this case there is a 0.8 probability that you have symptoms even though you don't have the virus, therefore the FPR rate of 0.05 multiplies this value as it is the probability that you will test positive even though you don't have the virus. 

```python
(0.9 * 0.2 + 0.05 * 0.8)*100
```

c) This is the same as above but for the other branch. We multiply the TPR by the positive branch and FPR by the negative branch. Remember that there are different TPRs and FPRs for this test. 0.8 * 0.0005 + 0.01 * 0.9995 = 0.01039 or 1.04%

```python
(0.8 * 0.0005 + 0.01 * 0.9995)*100
```

Here is Bayes' theorem: 



Using the above tree diagram, we treat P(S|V) and P(S|notV) as our priors and the right most branch as the likelihoods. 

Therefore, P(V|+S) = P(+|SV)xP(S|V)/(P(+|SV)xP(S|V) + P(+|S,notV)xP(S|notV))

= 0.9 x 0.2/(0.9 x 0.2 + 0.05 x 0.8) = 0.818 or 81.8%

```python
(0.9 * 0.2)/(0.9 * 0.2 + 0.8 * 0.05)
```

g) We know from the previous question that the probability this person has the virus after one test is 0.03848. This becomes our updated prior and replaces P(notS|V). Therefore, P(notS|notV) must equal 1 - this updated prior. 

Then, we just find the probability of testing positive in this branch using the updated priors. Testing positive is simply multiplying our prior by likelihood, and we have to do this and sum for both options of our likelihood with positive results.

It then becomes: updated prior x 0.8 + 1 - updated prior * 0.01 = 0.04 or 4%

```python
priorupdate*0.8 + (1-priorupdate)*0.01
```

(0.03848 x 0.8)/0.0404 = 0.7619 or 76.2%
